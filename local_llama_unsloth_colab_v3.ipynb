## 4. Carregar e validar o JSONL


```python
import json
from statistics import mean

def load_chat_jsonl(path):
    examples = []
    with open(path, encoding='utf-8') as file:
        for line_number, line in enumerate(file, start=1):
            if not line.strip():
                continue
            payload = json.loads(line)
            messages = payload.get('messages')
            if not isinstance(messages, list) or len(messages) < 3:
                raise ValueError(f'Linha {line_number} sem messages válidas')
            for message in messages:
                if message.get('role') not in {'system', 'user', 'assistant'}:
                    raise ValueError(f'Linha {line_number} com role inválido: {message}')
                if not message.get('content'):
                    raise ValueError(f'Linha {line_number} com content vazio: {message}')
            examples.append(payload)
    if not examples:
        raise ValueError('Dataset vazio')
    return examples
```

```python
examples = []

for filename in DATASET_FILES:
    file_path = DATA_DIR / filename
    try:
        file_examples = load_chat_jsonl(file_path)
        examples.extend(file_examples)
        print(f'Sucesso: {len(file_examples)} exemplos carregados de {filename}')
    except Exception as e:
        print(f'Erro ao carregar {filename}: {e}')

print(f'\n--> Total de exemplos combinados para o treinamento: {len(examples)}')
```

## 5. Carregar Llama com Unsloth

O modelo abaixo é carregado em 4 bits para reduzir VRAM. Você pode trocar por outro modelo compatível do Unsloth.


```python
from unsloth import FastLanguageModel

MODEL_NAME = 'unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit'
MAX_SEQ_LENGTH = 2048
LOAD_IN_4BIT = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=MODEL_NAME,
    max_seq_length=MAX_SEQ_LENGTH,
    dtype=None,
    load_in_4bit=LOAD_IN_4BIT,
)
```

## 6. Formatar com chat template e tokenizar preview

Aqui o dataset passa pelo tokenizer. O objetivo é medir tamanho em tokens e verificar se haverá truncamento.


```python
def render_messages(messages, tokenizer):
    return tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False,
    )

def format_dataset_into_model_input(examples, tokenizer):
    return [
        {
            'text': render_messages(example['messages'], tokenizer),
            'metadata': example.get('metadata', {}),
        }
        for example in examples
    ]
```

```python
formatted_examples = format_dataset_into_model_input(examples, tokenizer)

token_lengths = []
truncated = 0
for example in formatted_examples:
    tokens = tokenizer(
        example['text'],
        truncation=True,
        max_length=MAX_SEQ_LENGTH,
        add_special_tokens=False,
    )
    length = len(tokens['input_ids'])
    token_lengths.append(length)
    if length >= MAX_SEQ_LENGTH:
        truncated += 1

{
    'examples': len(formatted_examples),
    'min_tokens': min(token_lengths),
    'avg_tokens': round(mean(token_lengths), 2),
    'max_tokens': max(token_lengths),
    'truncated_examples': truncated,
}
```

## 7. Preparar LoRA


```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        'q_proj', 'k_proj', 'v_proj', 'o_proj',
        'gate_proj', 'up_proj', 'down_proj',
    ],
    lora_alpha=16,
    lora_dropout=0.0,
    bias='none',
    use_gradient_checkpointing='unsloth',
    random_state=3407,
    use_rslora=False,
    loftq_config=None,
)
```

## 8. Treinar

Para demonstração, `max_steps=150`. Para um treino mais completo, aumente esse valor e monitore perda/overfitting.

A saída do treinamento também usa `OUTPUT_DIR`, que já aponta para o Google Drive.


```python
from datasets import Dataset
from trl import SFTConfig, SFTTrainer
import inspect

train_dataset = Dataset.from_list(formatted_examples)

sft_params = inspect.signature(SFTConfig.__init__).parameters
sft_kwargs = {
    'output_dir': str(OUTPUT_DIR),
    'per_device_train_batch_size': 2,
    'gradient_accumulation_steps': 4,
    'warmup_steps': 5,
    'max_steps': 150,
    'learning_rate': 2e-4,
    'fp16': True,
    'logging_steps': 1,
    'optim': 'adamw_8bit',
    'weight_decay': 0.01,
    'lr_scheduler_type': 'linear',
    'seed': 3407,
    'report_to': 'none',
    'dataset_text_field': 'text',
    'packing': False,
    'padding_free': False
}

if 'max_length' in sft_params:
    sft_kwargs['max_length'] = MAX_SEQ_LENGTH
elif 'max_seq_length' in sft_params:
    sft_kwargs['max_seq_length'] = MAX_SEQ_LENGTH

sft_kwargs = {key: value for key, value in sft_kwargs.items() if key in sft_params}
sft_config = SFTConfig(**sft_kwargs)

trainer_params = inspect.signature(SFTTrainer.__init__).parameters
trainer_kwargs = {
    'model': model,
    'train_dataset': train_dataset,
    'args': sft_config
}

if 'processing_class' in trainer_params:
    trainer_kwargs['processing_class'] = tokenizer
elif 'tokenizer' in trainer_params:
    trainer_kwargs['tokenizer'] = tokenizer

trainer = SFTTrainer(**trainer_kwargs)
trainer.train()
```

## 11. Integração com LangChain (Requisito 2)

Esta seção encapsula o modelo customizado gerado no Unsloth dentro da arquitetura LangChain usando LCEL e um Custom LLM Wrapper para realizar consultas contextualizadas com o banco de dados hospitalar fictício.


```python
!pip install -U langchain langchain-core
```

```python
from typing import Any, List, Mapping, Optional
from langchain_core.language_models.llms import LLM
from langchain_core.callbacks.manager import CallbackManagerForLLMRun
from unsloth import FastLanguageModel

class UnslothLangChainLLM(LLM):
    """Wrapper customizado para integrar modelos otimizados pelo Unsloth ao ecossistema LangChain."""

    model: Any
    tokenizer: Any
    max_new_tokens: int = 512
    temperature: float = 0.2

    @property
    def _llm_type(self) -> str:
        return "unsloth-llama3"

    def _call(
        self,
        prompt: str,
        stop: Optional[List[str]] = None,
        run_manager: Optional[CallbackManagerForLLMRun] = None,
        **kwargs: Any,
    ) -> str:
        FastLanguageModel.for_inference(self.model)
        inputs = self.tokenizer([prompt], return_tensors="pt").to("cuda")

        outputs = self.model.generate(
            **inputs,
            max_new_tokens=self.max_new_tokens,
            temperature=self.temperature,
            do_sample=True
        )

        input_len = inputs.input_ids.shape[1]
        generated_tokens = outputs[0][input_len:]

        return self.tokenizer.decode(generated_tokens, skip_special_tokens=True)

# Inicializando o componente LangChain
llm_custom = UnslothLangChainLLM(model=model, tokenizer=tokenizer)
```

```python
import datetime

# Banco de dados em memória simulando o Prontuário Eletrônico do Hospital
# Aqui devemos buscar datasets reais, aninimizar e incluir aqui para consulta
MOCK_PATIENT_DATABASE = {
    "12345": {
        "nome": "Ana Maria da Silva",
        "id": 42,
        "historico_obstetrico": "G3P2A1 (3 gestações, 2 partos, 1 aborto)",
        "ultimo_papanicolau": "2023-04-12",
        "ultima_mamografia": "Nenhuma registrada",
        "historico_familiar": "Mãe diagnosticada com câncer de mama aos 45 anos",
        "medicamentos_ativos": ["Enalapril 10mg"]
    },
    "67890": {
        "nome": "Mariana Souza",
        "id": 26,
        "historico_obstetrico": "G0P0A0 (Nuligesta)",
        "ultimo_papanicolau": "2025-10-10",
        "ultima_mamografia": "Não se aplica",
        "historico_familiar": "Sem registros dignos de nota",
        "medicamentos_ativos": ["Anticoncepcional Oral Combinado"]
    }
}

def get_patient_context(patient_id: str) -> str:
    patient = MOCK_PATIENT_DATABASE.get(patient_id)
    if not patient:
        return "Paciente não encontrada no sistema hospitalar."

    hoje = datetime.date.today()
    ultimo_pap = datetime.datetime.strptime(patient["ultimo_papanicolau"], "%Y-%m-%d").date()
    anos_desde_pap = (hoje - ultimo_pap).days / 365.25

    alerta_preventivo = ""
    if anos_desde_pap > 3.0:
        alerta_preventivo = f" - ALERTA: Exame Papanicolau em atraso há {round(anos_desde_pap - 3, 1)} anos."

    contexto = (
        f"--- CONTEXTO ATUAL DA PACIENTE ---\n"
        f"Nome: {patient['nome']} | Idade: {patient['id']} anos\n"
        f"Histórico Obstétrico: {patient['historico_obstetrico']}\n"
        f"Histórico Familiar Relevante: {patient['historico_familiar']}\n"
        f"Medicamentos em Uso: {', '.join(patient['medicamentos_ativos'])}\n"
        f"Último Preventivo: {patient['ultimo_papanicolau']}{alerta_preventivo}\n"
        f"-----------------------------------"
    )
    return contexto
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import JsonOutputParser

output_parser = JsonOutputParser()

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "Você é um assistente de apoio à triagem em saúde e segurança da mulher. "
               "Analise o relato da paciente considerando o contexto clínico fornecido. "
               "Responda estritamente em um JSON válido contendo as seguintes chaves: "
               "'classificacao_risco', 'sinais_alerta', 'proxima_etapa_fluxo', 'resposta_acolhedora'.\n\n"
               "Instruções extras de segurança:\n"
               "- Se houver alerta de exames preventivos em atraso no contexto, adicione nos 'sinais_alerta' e mencione na 'resposta_acolhedora'.\n"
               "- Responda apenas com o objeto JSON estruturado, sem explicações fora dele."),
    ("user", "{contexto_clinico}\n\nRelato da Paciente: {relato_paciente}")
])

triagem_chain = prompt_template | llm_custom | output_parser
```

```python
import pprint

id_paciente = "12345"
contexto = get_patient_context(id_paciente)
relato = "Estou sentindo uma dor constante na boca do estômago e meu peito esquerdo parece que tem um ponto puxando quando levanto o braço."

resultado_triagem = triagem_chain.invoke({
    "contexto_clinico": contexto,
    "relato_paciente": relato
})

pprint.pprint(resultado_triagem)
```
